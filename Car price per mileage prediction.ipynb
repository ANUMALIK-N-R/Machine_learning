```python
car['engine_size'] =car['Engine Information.Engine Type'].str.extract(r'(\d+\.\d+)L')
car['cylinders'] = car['Engine Information.Engine Type'].str.extract(r'(\d+) cylinder')
car['engine_size'] = pd.to_numeric(car['engine_size'], errors='coerce')
car['cylinders'] = pd.to_numeric(car['cylinders'], errors='coerce')
```

```python
car['engine_size']=car['engine_size'].fillna(car['engine_size'].mean())
car['cylinders']=car['cylinders'].fillna(car['cylinders'].mode()[0])
```

```python
x=car.iloc[:,0:9].values
y=car.iloc[:,9:10].values
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_scaled=s.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,train_size=0.8,random_state=True)
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
y_pred=l.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
